```python
import pandas as pd
import scipy.stats as stats

# Load the collected data
pr_df = pd.read_csv('pr_metrics.csv')

# Separate merged and non-merged PRs
merged_prs = pr_df[pr_df['Is_Merged'] == True]
rejected_prs = pr_df[pr_df['Is_Merged'] == False]
```

```python
merged_prs
```

```python
from scipy.stats import shapiro

# Shapiro-Wilk test for normality
stat, p = shapiro(merged_prs['Time_to_Review'])
print('Accepted PRs Time_to_Review Normality p-value:', p)

stat, p = shapiro(rejected_prs['Time_to_Review'])
print('Rejected PRs Time_to_Review Normality p-value:', p)
```

```python
rejected_prs.describe
```

```python

def compare_metrics(metric):
    t_stat, p_value = stats.mannwhitneyu(merged_prs[metric], rejected_prs[metric])
    print(f"Metric: {metric}, t-statistic: {t_stat}, p-value: {p_value}")

# List of metrics to compare
metrics = ['Time_to_Review','Comments','Review_Comments','Changed_Files','Additions','Deletions','Total_Changes', 'CI_Status','Replicated_Code','Wont_Fix','Superseded']

for metric in metrics:
    compare_metrics(metric)
```
